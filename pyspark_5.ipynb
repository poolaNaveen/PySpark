```python
#Flatten
from pyspark.sql import SparkSession
from pyspark.sql.functions import explode, array
from pyspark.sql import functions as F
spark = SparkSession.builder.appName('flatten').getOrCreate()
data = [('rcb',),('csk',),('mi',),('kkr',)]
df = spark.createDataFrame(data,['team'])
df2 = df.select(explode(array(df.team)).alias('team'))
df2.show()
df1 = df.alias('t1').crossJoin(df.alias('t2'))\
    .where(F.col('t1.team') != F.col('t2.team'))\
    .select(F.col('t1.team').alias('team1'), F.col('t2.team').alias('team2'))
df1.show
df1.printSchema()
```
