```python
# Count
from pyspark.sql import SparkSession
from pyspark.sql import *
spark = SparkSession.builder.appName('day9').getOrCreate()
data = spark.sparkContext.parallelize([1,2,3,4,5,6])
result = data.reduce(lambda x,y:x+y)
print(result)
```

```python
from pyspark.sql import SparkSession
from pyspark.sql import *
spark = SparkSession.builder.appName('day9').getOrCreate()
data = spark.sparkContext.parallelize([
    (1,'naveen'),
    (2,'kamal'),
    (3,'babaulu'),
    (4,'lohith'),
    (1,'naveen'),
    (2,'kamal'),
    (3,'babaulu'),
    (4,'lohith'),
    (1,'naveen'),
    (2,'kamal'),
    (3,'babaulu'),
    (4,'lohith')

])
result = data.reduceByKey(lambda x,y: max(x,y))
print(result.collect())
```
